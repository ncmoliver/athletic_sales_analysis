```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv("./Resources/athletic_sales_2020.csv")
df_2021 = pd.read_csv("./Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], axis=0, join="inner",  keys=['2020', '2021'])
combined_df
```

```python
# Check if any values are null.
combined_df.isnull().sum()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
combined_df.columns
```

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
prod_sold_reg_state_cty = combined_df.groupby(['region', 'state', 'city'])[['units_sold']].sum()
total_prod_sold_reg_state_cty = prod_sold_reg_state_cty.rename(columns=({'units_sold' : 'Total_Products_Sold'}))

# Show the top 5 results.
total_prod_sold_reg_state_cty = total_prod_sold_reg_state_cty.sort_values(by='Total_Products_Sold', ascending=False)
total_prod_sold_reg_state_cty.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pvt_numSold_reg_state_cty = pd.pivot_table(combined_df,
               index=['region', 'state', 'city'],
               values= 'units_sold',
               aggfunc='count')

# Rename the "units_sold" column to "Total_Products_Sold"
pvt_numSold_reg_state_cty = pvt_numSold_reg_state_cty.rename(columns=({ 'units_sold' : 'Total_Products_Sold'}))

# Show the top 5 results.
pvt_numSold_reg_state_cty.sort_values(by='Total_Products_Sold', ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
combined_df.columns
```

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_df = combined_df.groupby(['region', 'state', 'city'])[['total_sales']].sum()
grouped_df = grouped_df.rename(columns=({'total_sales' : 'Total Sales'}))
                  
# Show the top 5 results.
top_sales_by_region = grouped_df.sort_values(by='Total Sales', ascending=False).head(5)
top_sales_by_region
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_by_region_s_c = combined_df.pivot_table(index=['region', 'state', 'city', 'product'],
                       values= 'total_sales',
                       aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
renamed_total_sales_df = total_sales_by_region_s_c.rename(columns=({'total_sales' : 'Total Sales'}))

# Show the top 5 results.
renamed_total_sales_df.sort_values(by='Total Sales', ascending=False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_df = combined_df.loc[combined_df['product'] == "Women's Athletic Footwear"]
womens_athletic_footwear_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_units_womens_athletic_footwear_df = womens_athletic_footwear_df.groupby(['retailer', 'region', 'state', 'city'])[['units_sold']].sum()
total_units_womens_athletic_footwear_df_renamed = total_units_womens_athletic_footwear_df.rename(columns=({'units_sold' : 'Womens_Footwear_Units_Sold'}))
# Show the top 5 results.
top_5 = total_units_womens_athletic_footwear_df_renamed.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_athletic_footwear_pivot = pd.pivot_table(womens_athletic_footwear_df,
                                                                        index=['retailer', 'region', 'state', 'city'],
                                                                        values='units_sold',
                                                                        aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
renamed_df = womens_athletic_footwear_pivot.rename(columns=({'units_sold' : 'Womens_Footwear_Units_Sold'}))

# Show the top 5 results.
top_5_units_sold_womens_athletic_footwear = renamed_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
top_5_units_sold_womens_athletic_footwear
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womens_athletic_footwear_pivot = pd.pivot_table(womens_athletic_footwear_df,
                                                index='invoice_date',
                                                values='total_sales',
                                                aggfunc='max')

# Optional: Rename the "total_sales" column to "Total Sales"
womens_athletic_footwear_pivot = womens_athletic_footwear_pivot.rename(columns=({'total_sales' : 'Total Sales'}))

# Show the table.
womens_athletic_footwear_pivot.sort_values(by='Total Sales', ascending=False).head(1)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
D_resampled_womens_athletic_footwear_pivot = womens_athletic_footwear_pivot.resample('D').sum()
D_resampled_womens_athletic_footwear_pivot = D_resampled_womens_athletic_footwear_pivot.rename(columns=({'Total Sales' : 'Total \'Daily\' Sales'}))
# Sort the resampled pivot table in descending order on "Total Sales".
sorted_resampled_womens_athletic_footwear_pivot= D_resampled_womens_athletic_footwear_pivot.sort_values(by='Total \'Daily\' Sales', ascending=False)
sorted_resampled_womens_athletic_footwear_pivot.head()
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
W_resampled_womens_athletic_footwear_pivot = womens_athletic_footwear_pivot.resample('W').sum()
D_resampled_womens_athletic_footwear_pivot = D_resampled_womens_athletic_footwear_pivot.rename(columns=({'Total Sales' : 'Total \'Daily\' Sales'}))
# Sort the resampled pivot table in descending order on "Total Sales".
W_sorted_resampled_womens_athletic_footwear_pivot= W_resampled_womens_athletic_footwear_pivot.sort_values(by='Total Sales', ascending=False).head(1)
W_sorted_resampled_womens_athletic_footwear_pivot
```
